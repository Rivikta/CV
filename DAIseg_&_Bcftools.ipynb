The first examples with no EM and posterior decoding algorithm is
```
python3 dai.seg.py \
    --obs_samples samples.txt \
    --bed file.bed \
    --HMM_par par.file.txt \
    --EM no \
    --prepared_file obs.txt \
    --o out_prefix \
    --arch_cover arch.covering.txt \
    --decoding posterior \
    --cut_off 0.9
```



````python
!python3 dai.seg.py \
    --obs_samples obserables.txt \
    --bed chr22.hg19.bed \
    --HMM_par par.file.txt \
    --EM no \
    --prepared_file chr22.txt \
    --o chr22_noEM \
    --arch_cover hg19.arch.covering.chr22.txt \
    --decoding posterior \
    --cut_off 0.9
````

The second example with EM on 10 samples and Viterbi algorithm


```
python3 dai.seg.py \
    --obs_samples samples.txt \
    --bed file.bed \
    --HMM_par par.file.txt \
    --EM yes \
    --EM_samples 10 \
    --EM_steps 20 \
    --prepared_file obs.txt \
    --o out_prefix \
    --arch_cover arch.covering.txt \
    --decoding posterior \
    --cut_off 0.9
```



````python
!python3 dai.seg.py \
    --obs_samples obserables.txt \
    --bed chr22.hg19.bed \
    --HMM_par par.file.txt \
    --EM yes \
    --EM_samples 10 \
    --EM_steps 20 \
    --prepared_file chr22.txt \
    --o chr22_EM \
    --arch_cover hg19.arch.covering.chr22.txt \
    --decoding posterior \
    --cut_off 0.9
````
